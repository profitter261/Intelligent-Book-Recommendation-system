```python
import re
import ast

def clean_and_join_genres(genre_info):
    if isinstance(genre_info, str):
        try:
            # Safely evaluate the string representation of the list
            genre_list = ast.literal_eval(genre_info)
        except (ValueError, SyntaxError):
            # If evaluation fails, assume it's not a list string and return as is
            return genre_info
    elif isinstance(genre_info, list):
        genre_list = genre_info
    else:
        return genre_info # Return as is if not a string or list

    cleaned_list = []
    if isinstance(genre_list, list):
        for genre in genre_list:
            if isinstance(genre, str):
                # Remove text within parentheses and the parentheses themselves
                cleaned_genre = re.sub(r'\(.*?\)', '', genre)
                # Remove leading and trailing whitespace
                cleaned_genre = cleaned_genre.strip()
                # Remove any leading non-alphabet characters
                cleaned_genre = re.sub(r'^[^a-zA-Z]*', '', cleaned_genre)
                if cleaned_genre:  # Only add if not empty after cleaning
                    cleaned_list.append(cleaned_genre)
    return ', '.join(cleaned_list)

df_clean['Extracted_Genre_Info_String'] = df_clean['Extracted_Genre_Info'].apply(clean_and_join_genres)
display(df_clean[['Extracted_Genre_Info', 'Extracted_Genre_Info_String']].head())
```

```python
import re

def extract_minutes(listening_time):
    if isinstance(listening_time, str) and listening_time != '-1':
        hours = 0
        minutes = 0
        if 'hours' in listening_time:
            hours_match = re.search(r'(\d+)\s*hours', listening_time)
            if hours_match:
                hours = int(hours_match.group(1))
        if 'minutes' in listening_time:
            minutes_match = re.search(r'(\d+)\s*minutes', listening_time)
            if minutes_match:
                minutes = int(minutes_match.group(1))
        return hours * 60 + minutes
    return 0

# Apply the function to the 'Listening Time' column
df['Listening_Time_Minutes'] = df['Listening Time'].apply(lambda x: extract_minutes(x))

display(df[['Listening Time', 'Listening_Time_Minutes']])
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.cluster import KMeans
from sklearn.metrics import precision_score, recall_score

# ==========================
# Load & Prepare Data
# ==========================
df = pd.read_csv("audible_data_cleaned.csv")

# Combine text features (title + description + genre)
df["features"] = (
    df["Book Name"].fillna("") + " " +
    df["Description"].fillna("") + " " +
    df["Genre"].fillna("")
)

# TF-IDF embeddings
vectorizer = TfidfVectorizer(stop_words="english", max_features=5000)
X = vectorizer.fit_transform(df["features"])

# ==========================
# 1. Content-Based Filtering
# ==========================
cosine_sim = cosine_similarity(X)

def recommend_content_based(book_index, top_k=5):
    scores = list(enumerate(cosine_sim[book_index]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)[1:top_k+1]
    return [df.iloc[i]["Book Name"] for i, _ in scores]

# ==========================
# 2. Clustering-Based Recommendations
# ==========================
kmeans = KMeans(n_clusters=10, random_state=42)
df["Cluster"] = kmeans.fit_predict(X)

def recommend_cluster_based(book_index, top_k=5):
    cluster = df.iloc[book_index]["Cluster"]
    cluster_books = df[df["Cluster"] == cluster].index.tolist()
    cluster_books = [i for i in cluster_books if i != book_index]
    return [df.iloc[i]["Book Name"] for i in cluster_books[:top_k]]

# ==========================
# 3. Hybrid Recommendations
# ==========================
def recommend_hybrid(book_index, top_k=5, alpha=0.5):
    # Content similarity
    content_scores = cosine_sim[book_index]

    # Cluster similarity (boost books in same cluster)
    cluster = df.iloc[book_index]["Cluster"]
    cluster_mask = (df["Cluster"] == cluster).astype(int).values
    hybrid_scores = alpha * content_scores + (1 - alpha) * cluster_mask

    scores = list(enumerate(hybrid_scores))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)[1:top_k+1]
    return [df.iloc[i]["Book Name"] for i, _ in scores]

# ==========================
# 4. Evaluation
# ==========================
# Dummy relevance assumption:
# Books in the same genre are considered "relevant"
def evaluate_model(recommender_fn, k=5, n_samples=100):
    precisions, recalls = [], []

    for idx in np.random.choice(len(df), n_samples, replace=False):
        true_genre = df.iloc[idx]["Genre"]
        recommended = recommender_fn(idx, top_k=k)
        relevant = df[df["Genre"] == true_genre]["Book Name"].tolist()

        hits = sum([1 for r in recommended if r in relevant])
        precision = hits / k
        recall = hits / min(len(relevant)-1, k) if len(relevant) > 1 else 0

        precisions.append(precision)
        recalls.append(recall)

    return np.mean(precisions), np.mean(recalls)

# Compare models
p_content, r_content = evaluate_model(recommend_content_based)
p_cluster, r_cluster = evaluate_model(recommend_cluster_based)
p_hybrid, r_hybrid = evaluate_model(recommend_hybrid)

print("Content-Based → Precision:", p_content, " Recall:", r_content)
print("Cluster-Based → Precision:", p_cluster, " Recall:", r_cluster)
print("Hybrid → Precision:", p_hybrid, " Recall:", r_hybrid)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
from sklearn.metrics import ndcg_score, average_precision_score
from sentence_transformers import SentenceTransformer
from sklearn.metrics import mean_squared_error
from scipy.sparse import csr_matrix
import torch # Import torch for tensor checks

# ==========================
# Load dataset
# ==========================
df = pd.read_csv("/content/cleaned_audible_data (1).csv")

# Combine features (title + description + genre)
df["features"] = (
    df["Book Name"].fillna("") + " " +
    df["Description"].fillna("") + " " +
    df["Genre"].fillna("")
)

# ==========================
# Handle single-occurrence genres for stratification
# ==========================
# Count the occurrences of each genre
genre_counts = df['Genre'].value_counts()
# Identify genres with only one occurrence
single_occurrence_genres = genre_counts[genre_counts == 1].index
# Remove rows with single-occurrence genres
df_filtered = df[~df['Genre'].isin(single_occurrence_genres)].copy()

# ==========================
# Train/Test Split
# ==========================
# Perform stratified split on the filtered DataFrame
train_df, test_df = train_test_split(df_filtered, test_size=0.2, random_state=42, stratify=df_filtered["Genre"])

# Reset index after splitting to ensure contiguous integer indexing
train_df = train_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)


# ----- TF-IDF -----
vectorizer = TfidfVectorizer(stop_words="english", max_features=5000)
X_train_tfidf = vectorizer.fit_transform(train_df["features"])
X_test_tfidf = vectorizer.transform(test_df["features"])

# ----- Sentence-BERT -----
bert_model = SentenceTransformer("all-MiniLM-L6-v2")  # lightweight & fast
X_train_bert = bert_model.encode(train_df["features"], convert_to_tensor=True)
X_test_bert = bert_model.encode(test_df["features"], convert_to_tensor=True)

# ==========================
# Evaluation Function
# ==========================
def evaluate_model(X_train, X_test, train_df, test_df, k_values=[5, 10], hybrid=False, alpha=0.7):
    results = {}
    for k in k_values:
        precisions, recalls, ndcgs, maps, rmses = [], [], [], [], []

        for idx in np.random.choice(test_df.index, min(100, len(test_df)), replace=False):
            true_genre = test_df.loc[idx, "Genre"]

            # Query vector
            if isinstance(X_test, (np.ndarray, csr_matrix)):
                q_vec = X_test[test_df.index.get_loc(idx)].reshape(1, -1)
                sims = cosine_similarity(q_vec, X_train).flatten()
            elif isinstance(X_test, torch.Tensor):
                q_vec = X_test[test_df.index.get_loc(idx)].unsqueeze(0)
                sims = cosine_similarity(q_vec.cpu(), X_train.cpu()).flatten()
            else:
                raise TypeError(f"Unsupported input type for X_test: {type(X_test)}")

            # Hybrid
            if hybrid:
                norm_ratings = train_df["Rating"] / train_df["Rating"].max()
                norm_reviews = train_df["Number of Reviews"] / train_df["Number of Reviews"].max()
                hybrid_boost = alpha * sims + (1 - alpha) * (0.7 * norm_ratings + 0.3 * norm_reviews)
                sims = hybrid_boost

            # Top-k recommendations
            top_indices = sims.argsort()[::-1][:k]
            recommended = train_df.iloc[top_indices]

            # Relevance vector (same genre = 1, else 0)
            relevant = (train_df["Genre"] == true_genre).astype(int).values
            y_true = relevant
            y_scores = sims

            # Precision / Recall
            hits = sum([1 for r in recommended["Book Name"] if r in train_df[train_df["Genre"] == true_genre]["Book Name"].tolist()])
            precision = hits / k
            recall = hits / min(len(train_df[train_df["Genre"] == true_genre]), k) if len(train_df[train_df["Genre"] == true_genre]) > 0 else 0
            precisions.append(precision)
            recalls.append(recall)

            # NDCG / MAP
            relevant_indices = train_df[train_df["Genre"] == true_genre].index
            y_true_filtered = np.zeros_like(y_true)
            y_true_filtered[relevant_indices] = 1
            if len(y_true_filtered) > 0 and len(y_scores) > 0:
                y_scores_filtered = sims[train_df.index]
                ndcgs.append(ndcg_score([y_true_filtered], [y_scores_filtered]))
                maps.append(average_precision_score(y_true_filtered, y_scores_filtered))
            else:
                ndcgs.append(0)
                maps.append(0)

            # --- RMSE: Compare predicted scores vs. actual ratings for recommended items ---
            true_ratings = recommended["Rating"].values
            pred_scores = sims[top_indices]

            if len(true_ratings) > 0:
                rmse = mean_squared_error(true_ratings, pred_scores)
                rmses.append(rmse)

        # Aggregate metrics
        results[f"Precision@{k}"] = np.mean(precisions)
        results[f"Recall@{k}"] = np.mean(recalls)
        results[f"NDCG@{k}"] = np.mean(ndcgs)
        results[f"MAP@{k}"] = np.mean(maps)
        results[f"RMSE@{k}"] = np.mean(rmses) if rmses else None

    return results
# ==========================
# Run Evaluations
# ==========================
print("TF-IDF Results:", evaluate_model(X_train_tfidf, X_test_tfidf, train_df, test_df, k_values=[5, 10]))
print("BERT Results:", evaluate_model(X_train_bert, X_test_bert, train_df, test_df, k_values=[5, 10]))
print("Hybrid Results:", evaluate_model(X_train_tfidf, X_test_tfidf, train_df, test_df, k_values=[5, 10], hybrid=True))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recommend_hybrid(book_index, X_train, X_test, train_df, test_df, alpha=0.7, beta=0.7, top_k=5):
    # Query vector (BERT)
    q_vec = X_test[book_index].unsqueeze(0)
    sims = cosine_similarity(q_vec.cpu(), X_train.cpu()).flatten()

    # Normalize popularity signals
    norm_ratings = train_df["Rating"] / train_df["Rating"].max()
    norm_reviews = train_df["Number of Reviews"] / train_df["Number of Reviews"].max()

    # Compute hybrid score
    popularity = beta * norm_ratings + (1 - beta) * norm_reviews
    hybrid_scores = alpha * sims + (1 - alpha) * popularity

    # Get top-k recommendations
    top_indices = hybrid_scores.argsort()[::-1][:top_k]
    return train_df.iloc[top_indices][["Book Name", "Author", "Genre", "Rating", "Number of Reviews"]]
```
